```python
import datetime
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 

payload = alice.historical_data(alice.get_instrument_by_symbol("NSE", "INFY-EQ"),datetime.datetime(2022, 9, 15),datetime.datetime(2022, 9, 20),HistoricalDataType.Minute)
df = pd.DataFrame.from_records(payload["result"])
df
```

```python
#Let's write the function for Simple Moving Average 

def plot_sma(stock_symbol, df, MA_period = 20):

  def sma(price, MA_period):
    mean_price = price.rolling(MA_period).mean()
    return np.round(mean_price, 3)

  df['moving_avg'] = sma(df["close"],MA_period)

  # plotting and visualization of simple moving average
  df['close'].plot(c = 'k', figsize = (20,10), lw = 2, fontsize = 12)
  df['moving_avg'].plot(c = 'b', figsize = (20, 10), lw = 1)

  # show plot
  plt.title('Simple Moving Average: {}'.format(stock_symbol), fontsize = 20)
  plt.ylabel('Price in INR(₹)',fontsize = 15 )
  plt.xlabel('Date', fontsize = 15 )
  plt.legend()
  plt.grid()
  plt.show()

plot_sma("INFY", df)
```
